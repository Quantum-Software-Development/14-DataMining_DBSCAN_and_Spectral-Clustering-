## ‚úåÔ∏è Compara√ß√£o K-Means e DBSCAN com Dataset 'Monkey'

```python
# Fun√ß√£o para plotar o gr√°fico k-distancia

def plot_k_distance_graph(X3, k):
    neigh = NearestNeighbors(n_neighbors=k)
    neigh.fit(X3)
    distances, _ = neigh.kneighbors(X3)
    distances = np.sort(distances[:, k-1])
    plt.figure(figsize=(6, 4))
    plt.plot(distances)
    plt.xlabel('Pontos')
    plt.ylabel(f' Dist√¢ncia entre os vizinhos mais pr√≥ximos')
    plt.title('Gr√°fico de K-dist√¢cia')
    plt.show()

# Plota gr√°fico k-distancia
plot_k_distance_graph(X3, k=5)
```

```python
# DBSCAN
dbscan = DBSCAN(eps=0.8 , min_samples=2)
dbscan = dbscan. fit_predict(X3)

# K-Means
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans = kmeans.fit_predict(X3)

# Visualize the results
fig , (a, b) = plt.subplots(1, 2, figsize=(12, 5))

a.scatter(X3['d1'], X3['d2'], c=dbscan , cmap='viridis')
a. set_title('DBSCAN (MONKEY) ')

b. scatter(X3['d1'], X3['d2'], c=kmeans , cmap='viridis')
b.set_title('K-MEANS (MONKEY)')
```

## DBSCAN Analysis and Visualization


Perform DBSCAN clustering on the `monkey.csv` dataset (`X3`) and visualize the results using the dark mode theme and turquoise color palette.


```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Initialize a DBSCAN model with eps=0.8 and min_samples=2
dbscan = DBSCAN(eps=0.8, min_samples=2)

# 2. Fit the DBSCAN model to the X3 DataFrame and obtain the cluster labels
clusters = dbscan.fit_predict(X3)

# 3. Create a scatter plot of the 'd1' and 'd2' columns from X3, using the obtained cluster labels to color the points.
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X3['d1'], y=X3['d2'], hue=clusters, palette=turquoise_palette, legend='full')

# 5. Set the title of the plot to 'DBSCAN Clustering on Monkey Dataset' and add appropriate labels for the x and y axes.
plt.title('DBSCAN Clustering on Monkey Dataset')
plt.xlabel('d1')
plt.ylabel('d2')

# 6. Display the plot.
plt.show()
```

### Explanation of DBSCAN Analysis and Visualization

This section details the application of the DBSCAN clustering algorithm to the `monkey.csv` dataset (`X3`) and the subsequent visualization of its results. The steps performed are:

1.  **DBSCAN Model Initialization**: An instance of the `DBSCAN` model from `sklearn.cluster` is created. Key parameters are set:
    *   `eps=0.8`: This defines the maximum distance between two samples for one to be considered as in the neighborhood of the other. Points within this distance are considered neighbors.
    *   `min_samples=2`: This defines the number of samples (or total weight) in a neighborhood for a point to be considered as a core point. Including the point itself.
    These parameters are crucial for DBSCAN to identify dense regions (clusters) and separate them from sparser regions (noise).

2.  **Model Fitting and Cluster Prediction**: The initialized `dbscan` model is then fitted to the `X3` DataFrame using `dbscan.fit_predict(X3)`. This method not only fits the model but also immediately returns an array of cluster labels for each data point in `X3`. Points labeled with `-1` are considered noise.

3.  **Visualization Setup**: A new figure for plotting is created with a specified size (`figsize=(8, 6)`) to ensure good readability.

4.  **Scatter Plot Generation**: A scatter plot is generated using `seaborn.scatterplot`. The `d1` and `d2` columns of `X3` represent the x and y coordinates, respectively. The `hue=clusters` argument ensures that each cluster (and noise points) is colored differently based on the `clusters` array. Importantly, `palette=turquoise_palette` applies the custom turquoise color scheme defined in the initial setup, ensuring visual consistency with the dark mode theme.

5.  **Plot Customization**: The plot is given a descriptive title, 'DBSCAN Clustering on Monkey Dataset', and the x and y axes are labeled 'd1' and 'd2' respectively. This enhances the interpretability of the visualization.

6.  **Display Plot**: Finally, `plt.show()` displays the generated scatter plot, allowing for a visual assessment of the DBSCAN algorithm's performance on the `monkey.csv` dataset with the specified parameters and theme.

## Data Analysis Report: DBSCAN vs. K-Means on Monkey Dataset

### üá¨üáß English Report

#### Introduction
This report details the application and comparison of two popular clustering algorithms, DBSCAN (Density-Based Spatial Clustering of Applications with Noise) and K-Means, on the `monkey.csv` dataset. The analysis aims to understand how each algorithm groups data points based on their spatial distribution, particularly in a dataset that exhibits non-spherical clusters. All visualizations are presented with a dark mode theme and a custom turquoise color palette for improved readability and aesthetic consistency.

#### DBSCAN Clustering Analysis
DBSCAN was applied with `eps=0.8` and `min_samples=2`. The visualization clearly shows that DBSCAN effectively identifies the intricate, non-spherical clusters present in the `monkey.csv` dataset. The algorithm successfully groups the points into distinct, elongated shapes, characteristic of its ability to discover clusters of arbitrary shapes. Furthermore, DBSCAN correctly identifies and labels a significant number of points as noise (typically represented by a distinct color, often black or grey, in the visualizations), which are points that do not belong to any dense region. This ability to handle noise and discover complex cluster shapes is a key strength of DBSCAN, particularly evident in this dataset.

#### K-Means Clustering Analysis
K-Means was applied with `n_clusters=2` and `random_state=42`. As observed in the visualization, K-Means struggles to accurately cluster the `monkey.csv` dataset. Due to its inherent assumption of spherical clusters and its reliance on centroids, K-Means attempts to divide the data into two roughly equal-sized, convex regions. This results in misclassifications, where points belonging to the natural, intertwined clusters are incorrectly assigned to different groups. The output shows two large, somewhat overlapping clusters that do not reflect the true underlying structure of the data.

#### Comparison and Conclusion
The side-by-side comparison unequivocally demonstrates the superior performance of DBSCAN over K-Means for the `monkey.csv` dataset. DBSCAN's density-based approach allows it to delineate the complex, non-linear patterns present in the data, effectively separating the 'monkey' shape into its constituent parts and identifying outliers. In contrast, K-Means' centroid-based, spherical assumption leads to an inaccurate and less meaningful partitioning of the data. This highlights that the choice of clustering algorithm is highly dependent on the intrinsic geometry and density distribution of the dataset. For datasets with arbitrary shapes and varying densities, DBSCAN proves to be a more robust and appropriate choice.

# üë©üèª‚ÄçüöÄ Introduction to Spectral Clustering üõ∏

### Subtask:
Introduce Spectral Clustering as another algorithm suitable for non-spherical clusters and explain its general approach.

### Explanation of Spectral Clustering

Spectral Clustering is a sophisticated clustering technique particularly well-suited for identifying **non-spherical clusters** and clusters with **complex geometries**, much like those observed in the `monkey.csv` dataset. Unlike traditional algorithms such as K-Means, which struggle with such structures due to their reliance on centroid-based assignments and assumptions of convex shapes, Spectral Clustering operates on a different principle.

Its general approach involves transforming the clustering problem into a graph partitioning problem. Here's a breakdown of its core steps:

1.  **Construct a Similarity Graph**: The first step is to build a similarity graph (also known as an affinity matrix) from the data. Each data point represents a node in the graph, and edges between nodes are weighted by the similarity (or affinity) between the corresponding data points. Common similarity measures include Gaussian similarity (RBF kernel) or k-nearest neighbors.

2.  **Compute the Graph Laplacian**: From the similarity graph, a graph Laplacian matrix is derived. This matrix is crucial because its eigenvectors contain information about the structure of the graph.

3.  **Dimensionality Reduction**: The eigenvectors corresponding to the smallest eigenvalues of the Laplacian matrix are then selected. These eigenvectors form a new, lower-dimensional embedding of the data points. This step effectively performs **dimensionality reduction**, transforming the original data into a space where clusters are more linearly separable.

4.  **Clustering in the Embedded Space**: Finally, a standard clustering algorithm, typically **K-Means**, is applied to these new, embedded data points. Because the data has been transformed into a space where clusters are more convex and separable, K-Means can now effectively identify the clusters that were originally non-spherical in the high-dimensional space.

This two-step process‚Äîembedding the data into a lower-dimensional space using spectral methods and then applying K-Means‚Äîallows Spectral Clustering to overcome the limitations of distance-based algorithms on complex data distributions.

## Comparative Visualization of Clustering Algorithms


Create a combined visualization using subplots to display the clustering results of DBSCAN, K-Means, and Spectral Clustering side-by-side on the `X3` dataset, maintaining the dark mode theme and turquoise color palette.
